```
import pandas as pd
import numpy as np
from scipy.linalg import svd,norm,inv
import math
from time import process_time
```

```
train_X = np.load("ML-1M/train1.npy")
test_X = np.load("ML-1M/test1.npy")
```

```
losses =[]
times=[]
for i in range(1,6):
    train_X = np.load("ML-1M/train"+str(i)+".npy")
    test_X = np.load("ML-1M/test"+str(i)+".npy")
    train_X[train_X == 0] = np.nan
    row_means= np.nanmean(train_X, axis=1)
    t1 = process_time()
    #get average ratings per movie
    means = np.nanmean(train_X, axis=0)
    for i in range(train_X.shape[0]):
        for j in range(train_X.shape[1]):
            if np.isnan(train_X[i][j]):
                train_X[i][j] = means[j]

    #normalization
    for i in range(train_X.shape[0]):
        train_X[i] =train_X[i]-row_means[i]
    train_X

    U, s, VT = svd(train_X)
    Sigma= np.diag(s)
    k = 14
    Sigma = Sigma[:k, :k]
    U = U[:, :k]
    VT = VT[:k, :]

    t2 = process_time()
    print(t2-t1)
    times.append(t2-t1)
    def predict(row,product):
        pred = row_means[row]+U[row].dot(Sigma).dot(VT[:,product])
        return pred
        
    test_row, test_col = np.nonzero(test_X)
    rmse = 0
    for i in range(len(test_row)):
        pred = predict(test_row[i],test_col[i])
        true = test_X[test_row[i],test_col[i]]
        rmse += (pred-true)**2
    rmse /=len(test_row)
    rmse = rmse **0.5
    print(rmse)
    losses.append(rmse)
```

```
rmse = np.array(losses)
times = np.array(times)
```

```
print(rmse.mean())
print(2*rmse.std()/5**0.5)
print(times.mean())
print(2*times.std()/5**0.5)
```
